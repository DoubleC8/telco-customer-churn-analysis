```python
import numpy as np
import pandas as pd

# reading in file and creating a dataframe for the data in that file
df = pd.read_csv('Telco-Customer-Churn.csv')

# sanity check to make sure my file was imported correctly 
df.head()
```

```python
# checking data type of TotalCharges
print(df['TotalCharges'].info())
```

```python
# checking for any blank cells in TotalCharges
df[df['TotalCharges'] == ' ']
```

```python
# after checking, we found some blank entries
# these entries were due to specific customers having a tenure of 0
# so this means they literally just signed up
# instead of deleting these rows, I will assign all customers with tenure = 0 AND TotalCharges = " " to have a TotalCharges = 0 instead
df.loc[(df['tenure'] == 0) & (df['TotalCharges'] == ' '), 'TotalCharges'] = 0

# after doing this, I will change all values in this column to be numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])

# sanity check
df[df['TotalCharges'] == ' ']
print(df['TotalCharges'].info())
```

```python
# i will change all values in my churn columns
# Yes -> 1, No -> 0
df['Churn'] = np.where(df['Churn'] == 'Yes',  1, 0)

# sanity check
df.head()
```

```python
# creating new column Tenure_Cohort
# will do the following grouping:
# 0-12 Months | 13-24 Months | 25-48 Months | Over 48 Months

bins = [0, 12, 24, 48, np.inf]
labels = ['0-12 Months', '13-24 Months', '25-48 Months', 'Over 48 Months']

Tenure_Cohort = pd.cut(df['tenure'], bins = bins, labels = labels)
print(Tenure_Cohort)

df['Tenure_Cohort'] = Tenure_Cohort
df.head()
```

```python
df_copy = df.copy()
df_copy = df_copy.drop(columns='customerID')
df_copy.head()
```

```python
# exporting cleaned data
df_copy.to_csv('cleaned_telco_customer_churn.csv', index=False)
```

```python
import sqlite3

# creating a temp SQL db in my notebook
conn = sqlite3.connect('churn_analysis.db')

# loading in cleaned df into the SQL db
df_copy.to_sql('customers', conn, if_exists='replace', index=False)
```

```python
# writting my SQL queries
# getting the churn average for every customer
query = """
SELECT 
    (ROUND(AVG(Churn) * 100, 2) || '%') AS churn_avg
FROM customers
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
# calculating average by contract type
query = """
SELECT
    Contract, 
    COUNT(*) AS total_customers, 
    (ROUND(AVG(Churn) * 100, 2) || '%') AS churn_avg
FROM
    customers
GROUP BY 
    Contract
ORDER BY 
    AVG(Churn) DESC
"""

pd.read_sql(query, conn)
```

```python
# calculating average by Tenure_Cohort
# we want to see if the longer a customer stays, the less likely they are to cancel.
query = """
SELECT
    Tenure_Cohort, 
    COUNT(*) AS total_customers,
    (ROUND(AVG(Churn) * 100, 2) || '%') AS churn_avg
FROM
    customers
GROUP BY 
    Tenure_Cohort
ORDER BY 
    AVG(Churn) DESC
"""

pd.read_sql(query, conn)
```

```python
# writting a query to look at the highest-risk group (Month-to-Month users) 
# and see if having TechSupport makes them stay longer.
query = """
SELECT 
    Contract,
    TechSupport, 
    COUNT(*) AS total_customers, 
    (ROUND(AVG(Churn) * 100, 2) || '%') AS churn_avg
FROM 
    customers
WHERE
    Contract = 'Month-to-month'
GROUP BY 
    TechSupport
ORDER BY 
    AVG(Churn) DESC
"""

pd.read_sql(query, conn)
```
